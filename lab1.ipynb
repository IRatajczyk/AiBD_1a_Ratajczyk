```python
#import of necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Zadanie 3.
Zdefiniuj poniższą funkcję i sporządź jej wykres dla argumentów z danych przedziałów:

$f(x)=x^2+5$

* $-1<x<1$

* $-6<x<6$

* $0<x<5$


```python
#definicja funkcji jako prostej funkcji lambda
fun = lambda x : x*x+5

#inicjalizacja dziedzin kolejnych funkcji
domain_1 = np.linspace(-1,1)
domain_2 = np.linspace(-6,6)
domain_3 = np.linspace(0,5)

#tworzenie subplotów
f, (ax1,ax2,ax3) = plt.subplots(1,3,figsize=(30,10))
f.suptitle('Comprehensive comparison exemplary plots')

ax1.plot(domain_1,fun(domain_1))
ax1.set_title('First domain')
ax1.set_xlim([-1,1])
ax1.set_ylabel('y')
ax1.set_xlabel('x')

ax2.plot(domain_2,fun(domain_2))
ax2.set_title('Second domain')
ax2.set_xlim([-6,6])
ax2.set_ylabel('y')
ax2.set_xlabel('x')

ax3.plot(domain_3,fun(domain_3))
ax3.set_title('Third domain')
ax3.set_xlim([0, 5])
ax3.set_ylabel('y')
ax3.set_xlabel('x')


plt.show()
```
